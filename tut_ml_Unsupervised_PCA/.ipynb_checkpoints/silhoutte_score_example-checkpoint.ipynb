```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances

# Step 1: Generate 3 clusters with 2 integer points each in the range (0, 20)
np.random.seed(42)  # For reproducibility

cluster_centers = np.random.randint(0, 20, (3, 2))  # 3 cluster centers (integer values)
clusters = []
for center in cluster_centers:
    cluster_points = center + np.random.randint(-3, 3, (2, 2))  # 2 points per cluster, small variation
    clusters.append(cluster_points)

clusters = np.vstack(clusters)  # Convert list of arrays to a single array

# Step 2: Assign cluster labels (0, 1, 2) and names
labels = np.repeat([0, 1, 2], 2)  # Each cluster has 2 points
cluster_names = {0: "Red", 1: "Green", 2: "Blue"}
cluster_colors = {"Red": "red", "Green": "green", "Blue": "blue"}

# Step 3: Print all data points with their cluster names
print("Data Points and Their Clusters:")
for i, point in enumerate(clusters):
    print(f"Point {point} -> Cluster {cluster_names[labels[i]]}")

# Step 4: Select a random point for Silhouette Score calculation
random_index = np.random.choice(len(clusters))
random_point = clusters[random_index]
random_label = labels[random_index]
random_cluster_name = cluster_names[random_label]

# Step 5: Calculate Silhouette Score manually
distances = pairwise_distances(clusters)

# Compute 'a' (intra-cluster distance) -> Average distance to all points in the same cluster
same_cluster_mask = (labels == random_label) & (np.arange(len(clusters)) != random_index)
a = np.mean(distances[random_index, same_cluster_mask]) if np.any(same_cluster_mask) else 0

# Compute 'b' (inter-cluster distance) -> Minimum average distance to all points in other clusters
b_values = {}
for other_label in set(labels) - {random_label}:
    other_cluster_mask = labels == other_label
    avg_distance = np.mean(distances[random_index, other_cluster_mask])
    b_values[other_label] = avg_distance

# Find the nearest cluster for 'b'
nearest_cluster_label = min(b_values, key=b_values.get)
b = b_values[nearest_cluster_label]

# Compute silhouette score for the random point
silhouette_score = (b - a) / max(a, b)

# Step 6: Plot the clusters
plt.figure(figsize=(8, 6))
for i in range(3):
    cluster_points = clusters[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                color=cluster_colors[cluster_names[i]], label=cluster_names[i])

# Highlight the randomly chosen point
plt.scatter(random_point[0], random_point[1], color="black", edgecolors="yellow", s=200, label="Chosen Point", zorder=3)

# Step 7: Draw lines to other cluster points and annotate distances
for i, point in enumerate(clusters):
    if labels[i] != random_label:  # Only draw to other clusters
        dist = np.linalg.norm(random_point - point)  # Euclidean distance
        is_nearest = labels[i] == nearest_cluster_label  # Bold if nearest cluster

        line_style = 'k-' if is_nearest else 'k--'  # Solid for nearest, dashed otherwise
        line_width = 2.5 if is_nearest else 1.5  # Thicker line for nearest

        plt.plot([random_point[0], point[0]], [random_point[1], point[1]], 
                 line_style, linewidth=line_width, alpha=0.7)

        # Annotate distance on the plotted lines
        mid_x, mid_y = (random_point[0] + point[0]) / 2, (random_point[1] + point[1]) / 2
        plt.text(mid_x, mid_y, f"{dist:.2f}", fontsize=10, ha='center', color='black', fontweight="bold")

# Plot settings
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Clusters with Silhouette Score Calculation")
plt.legend()
plt.grid(True)
plt.show()

# Display results
print("\nSilhouette Score Calculation:")
print(f"Randomly chosen point: {random_point}")
print(f"Cluster name of chosen point: {random_cluster_name}")
print(f"Mean intra-cluster distance (a): {a:.4f}")
print(f"Mean nearest inter-cluster distance (b): {b:.4f}")
print(f"Silhouette score for the point: {silhouette_score:.4f}")
```
